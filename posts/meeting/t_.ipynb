```python
from typing import List
class Solution:
    def fullJustify(self, words: List[str], maxWidth: int) -> List[str]:
        i, j = 0, 0 
        cur_sum = 0
        ret = []
        while i < len(words) and j < len(words):
            cur_sum += len(words[j])
            
            if (cur_sum + (j-i))  > maxWidth:
                cur_sum -= len(words[j])
                num_word = j - i
                if num_word == 1:
                    first_space = maxWidth - cur_sum
                else:
                    average_space = (maxWidth - cur_sum) //(num_word-1) 
                    first_space = (maxWidth - cur_sum) % (num_word-1) + average_space 
                    other_space = average_space
                
                the_line = ""
                for k in range(i, j):
                    if k == i:
                        the_line = the_line + words[k] + (" " * first_space)
                    elif k == j-1:
                        the_line = the_line + words[k]
                    else:
                        the_line = the_line + words[k] + (" " * other_space)
                ret.append(the_line)

                cur_sum = 0
                i = j
            else:
                j += 1
        
        the_line = ""
        for k in range(i, j):
            if k != j-1:
                the_line = the_line + words[k] + " "
            else:
                the_line = the_line + words[k] + " " * (maxWidth - cur_sum - (j-i-1))
        ret.append(the_line)

        return ret
s = Solution()
words = ["Science","is","what","we","understand","well","enough","to","explain","to","a","computer.","Art","is","everything","else","we","do"]
maxWidth = 20
print(s.fullJustify(words, maxWidth))
```

```python
class Solution(object):
    def threeSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        sorted_num = sorted(nums)
        ret = []
        for i in range(len(sorted_num)):
            target = 0 - sorted_num[i]
            j, k = 0, len(sorted_num)-1
            while j < k:
                if j == i :
                    j += 1
                    continue
                if k == i:
                    k -= 1
                    continue
                if sorted_num[j] + sorted_num[k] < target:
                    j += 1
                elif sorted_num[j] + sorted_num[k] > target:
                    k -= 1
                else:
                    ret.append([sorted_num[i], sorted_num[j], sorted_num[k]])
        return ret

        # -4 -1 -1 0 1 2
s = Solution()
s.threeSum([-1,0,1,2,-1,-4])
```

```python
from typing import List
class Solution:
    def minCost(self, n: int, cuts: List[int]) -> int:
        cuts.sort()
        cuts = [0] + cuts + [n]

        def dfs(i: int, j: int) -> int:
            if i + 1 == j:  # 无需切割
                return 0
            # 枚举切割位置 cuts[k]
            print(range(i + 1, j))
            return min(dfs(i, k) + dfs(k, j) for k in range(i + 1, j)) + cuts[j] - cuts[i]

        return dfs(0, len(cuts) - 1)


s = Solution()
print(s.minCost(7, [1,3,4,5]))
```

```python
class Solution:
    def countKConstraintSubstrings(self, s: str, k: int) -> int:
        n = len(s)
        cnt = 0
        def check(ss, k):
            num_of_one = sum([int(i) for i in s])
            num_of_zero = len(ss) - num_of_one
            if num_of_one <= k or num_of_zero <= k:
                return True
            else:
                return False
        for i in range(1, n+1):
            for start in range(n-i+1):
                sub_s = s[start:start+i]
                if check(sub_s, k):
                    cnt += 1
        return cnt
s = Solution()
s.countKConstraintSubstrings("1010101", 1)
```

```python
from typing import List
class Solution:
    def countKConstraintSubstrings(self, s: str, k: int, queries: List[List[int]]) -> List[int]:
        ret = []
        # this_cnt = 0
        left, p = 0, 0 
        n_zero, n_one = 0, 0
        for i, c in enumerate(s):
            if c == "0":
                n_zero += 1
            elif c == "1":
                n_one += 1
            while n_zero > k and n_one > k:
                if s[left] == '0':
                    n_zero -= 1
                elif s[left] == '1':
                    n_one -= 1
                left += 1
            # this_cnt = i - left + 1
            ret.append(left)
        final_ret = []
        for i, j in queries:
            tmp_ret = 0
            for k in range(j, i-1, -1):
                tmp_ret += (k-max(ret[k], i)+1) 
            final_ret.append(tmp_ret)
        return final_ret
s = Solution()
print(s.countKConstraintSubstrings("010101", 1, [[0,5],[1,4],[2,3]]))
```

```python
class UnionFind:
    def __init__(self):
        """
        记录每个节点的父节点
        """
        self.father = {}
        self.num_of_sets = 0
        
    def find(self,x):
        """
        查找根节点
        路径压缩
        """
        if self.father[x] != None:
            self.father[x] = self.find(self.father[x])
            return self.father[x]
        else:
            return x
    
    def merge(self,x,y):
        """
        合并两个节点
        """
        root_x,root_y = self.find(x),self.find(y)
        
        if root_x != root_y:
            self.father[root_x] = root_y
            self.num_of_sets -= 1

    def is_connected(self,x,y):
        """
        判断两节点是否相连
        """
        return self.find(x) == self.find(y)
    
    def add(self,x):
        """
        添加新节点
        """
        if x not in self.father:
            self.father[x] = None
            self.num_of_sets += 1
            
from typing import List
class Solution:
    def findCircleNum(self, M: List[List[int]]) -> int:
        uf = UnionFind()
        for i in range(len(M)):
            uf.add(i)
        for i in range(len(M)):
            for j in range(i+1,len(M)):
                if M[i][j] == 1:
                    uf.merge(i,j)
    
        return uf.num_of_sets
    
M = [[1,1,0],[1,1,0],[0,0,1]]
s = Solution()
s.findCircleNum(M)
```

```python
class Solution:
    def shortestDistanceAfterQueries(self, n: int, queries: List[List[int]]) -> List[int]:
        f = list(range(n))
        from_p = [[] for i in range(n)]
        ans = [0] * len(queries)
        for i, (l, r) in enumerate(queries):
            from_p[r].append(l)
            if f[r] > f[l] + 1:
                f[r] = f[l] + 1
                for j in range(r + 1, n):
                    f[j] = (f[j - 1] + 1)
                    for k in from_p[j]:
                        f[j] = min(f[j], f[k] + 1)
            ans[i] = f[n - 1]
        return ans
s = Solution()
s.shortestDistanceAfterQueries(5, [[2,4],[0,2],[0,4]])
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Set high quality settings
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.dpi'] = 150
plt.rcParams['savefig.dpi'] = 300

# Simulate test data
np.random.seed(42)

# Desktop grasping scenario simulated data
desktop_pi0_times = np.random.normal(0.6, 0.3, 20)  # seconds
desktop_pi0_accuracy = np.random.normal(0.8, 0.15, 20)
desktop_pi0_accuracy = np.clip(desktop_pi0_accuracy, 0, 1)

desktop_act_times = np.random.normal(1.3, 0.8, 20)
desktop_act_accuracy = np.random.normal(0.95, 0.05, 20)
desktop_act_accuracy = np.clip(desktop_act_accuracy, 0, 1)

# Conveyor belt scenario simulated data
conveyor_pi0_times = np.random.normal(0.8, 0.2, 18)
conveyor_pi0_accuracy = np.random.normal(0.95, 0.05, 18)
conveyor_pi0_accuracy = np.clip(conveyor_pi0_accuracy, 0, 1)

conveyor_act_times = np.random.normal(1.4, 0.3, 18)
conveyor_act_accuracy = np.random.normal(0.6, 0.2, 18)
conveyor_act_accuracy = np.clip(conveyor_act_accuracy, 0, 1)

# Create compact boxplot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Response time boxplot
response_data = [desktop_pi0_times, desktop_act_times, 
                conveyor_pi0_times, conveyor_act_times]
labels = ['Desktop\nPI0', 'Desktop\nACT', 'Conveyor\nPI0', 'Conveyor\nACT']

box1 = ax1.boxplot(response_data, labels=labels, patch_artist=True,
                   boxprops=dict(facecolor='#2E86AB', alpha=0.7),
                   medianprops=dict(color='#C73E1D', linewidth=2),
                   whiskerprops=dict(color='#2E86AB', linewidth=1.5),
                   capprops=dict(color='#2E86AB', linewidth=1.5))

ax1.set_title('Response Time Distribution by Scenario & Model', fontsize=12, fontweight='bold')
ax1.set_ylabel('Response Time (seconds)', fontsize=10)
ax1.grid(axis='y', alpha=0.3)
ax1.tick_params(axis='x', labelsize=9)

# Accuracy boxplot
accuracy_data = [desktop_pi0_accuracy, desktop_act_accuracy, 
                conveyor_pi0_accuracy, conveyor_act_accuracy]

box2 = ax2.boxplot(accuracy_data, labels=labels, patch_artist=True,
                   boxprops=dict(facecolor='#A23B72', alpha=0.7),
                   medianprops=dict(color='#C73E1D', linewidth=2),
                   whiskerprops=dict(color='#A23B72', linewidth=1.5),
                   capprops=dict(color='#A23B72', linewidth=1.5))

ax2.set_title('Accuracy Distribution by Scenario & Model', fontsize=12, fontweight='bold')
ax2.set_ylabel('Accuracy (0-1)', fontsize=10)
ax2.grid(axis='y', alpha=0.3)
ax2.tick_params(axis='x', labelsize=9)

# Add mean value annotations
for i, data in enumerate(response_data):
    mean_val = np.mean(data)
    ax1.text(i+1, mean_val, f'{mean_val:.2f}s', ha='center', va='bottom', 
             fontsize=8, fontweight='bold', bbox=dict(boxstyle='round,pad=0.2', facecolor='white', alpha=0.8))

for i, data in enumerate(accuracy_data):
    mean_val = np.mean(data)
    ax2.text(i+1, mean_val, f'{mean_val:.2f}', ha='center', va='bottom', 
             fontsize=8, fontweight='bold', bbox=dict(boxstyle='round,pad=0.2', facecolor='white', alpha=0.8))

plt.tight_layout()
plt.savefig('PI0_model_stability_analysis.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```
